```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("dataset/ecommerce_sales.csv")
print(df)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Total revenue by product category.
category_revenue=df.groupby("Category")["TotalAmount"].sum()
category_revenue.plot(kind="bar")
plt.title("Revenue Generated by Product Category")
plt.xlabel("Categories")
plt.ylabel("Revenue")
plt.show()
```

```python
# Top 5 Products by Quantity Sold.
top_products_quantity=df.groupby("Product")["Quantity"].sum().head(5).sort_values(ascending=False)
top_products_quantity.plot(kind="bar",color="green")
plt.title("Top 5 Products by Quantity Sold")
plt.xlabel("Products")
plt.ylabel("Quantity")
plt.show()
```

```python
# Revenue by City
city_revenue=df.groupby("City")["TotalAmount"].sum()
city_revenue.plot(kind="bar")
plt.title("Revenue Generated by City")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.show()
```

```python
# Top 5 Customers by Spending
top_customers_spending=df.groupby("CustomerName")["TotalAmount"].sum().head(5).sort_values(ascending=False)
top_customers_spending.plot(kind="bar")
plt.title("Top 5 Customers by Total Spending")
plt.xlabel("Customers")
plt.ylabel("TotalSpent")
plt.show()
```
